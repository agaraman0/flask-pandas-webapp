```python
def nearestNaturalnum(row):
    if row[1] > 0:
        if row[1] - int(row[1]) > 0.5:
            return int(row[1]) + 1
        else:
            return int(row[1])
    else:
        if row[1] - int(row[1]) > -0.5:
            return int(row[1])
        else:
            return int(row[1]) - 1
```

```python
data['Retention Time RoundOff'] = data.apply(nearestNaturalnum,axis=1)
```

```python
data['RT1'] = data.apply(lambda row: int(row[1])+1 if row[1] - int(row[1]) > 0.5 else int(row[1]), axis = 1) 
```
